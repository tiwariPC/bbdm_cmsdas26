```python
import pickle
from pathlib import Path

pkl_path = Path("output/output_2017_full.pkl")
with open(pkl_path, "rb") as f:
    results = pickle.load(f)

print("type:", type(results))
print("n_datasets:", len(results))
print("dataset names (sample):", sorted(results.keys())[:15])
```

```python
name = sorted(results.keys())[0]
acc = results[name]
print("example dataset:", name)
print("accumulator keys:", list(acc.keys()))
print("cutflow:", dict(acc["cutflow"]))
```

```python
for key in acc:
    if key == "cutflow":
        continue
    ha = acc[key]
    h = getattr(ha, "_hist", ha)
    axes_info = [f"{a.name}: {a.label}" for a in h.axes]
    print(f"{key}: {type(h).__name__}")
    print(f"  axes: {axes_info}")
```

```python
h_met = getattr(acc["met_sr"], "_hist", acc["met_sr"])
edges = h_met.axes[0].edges
print("met_sr bin edges (first/last):", edges[:5], "...", edges[-3:])
```

```python
all_cf_keys = set()
for ds, a in results.items():
    all_cf_keys.update(dict(a["cutflow"]).keys())
print("cutflow keys (all datasets):", sorted(all_cf_keys))
```

```python

import matplotlib.pyplot as plt
import numpy as np


def is_data(name: str) -> bool:
    return ("Run2017" in name) or name.startswith("MET_")

# --- cutflow key order (edit to match your processor) ---
# Inspect once: sorted({k for acc in results.values() for k in acc["cutflow"].keys()})
CUT_KEYS = ["trigger", "met_pf_calo", "presel", "signal_region"]

datasets = sorted(results.keys())
data_names = [d for d in datasets if is_data(d)]
bkg_names = [d for d in datasets if not is_data(d)]

# optional: only some backgrounds (else use all MC)
# bkg_names = [n for n in bkg_names if n.startswith(("TT", "ZJets", "WJets", "DY"))]

def cf(ds, key):
    return int(dict(results[ds]["cutflow"]).get(key, 0))

# --- plot ---
fig, ax = plt.subplots(figsize=(9, 5))
x = np.arange(len(CUT_KEYS))
width = 0.65

bottom = np.zeros(len(CUT_KEYS))
for name in bkg_names:
    h = np.array([cf(name, k) for k in CUT_KEYS], dtype=float)
    ax.bar(x, h, width, bottom=bottom, label=name[:40] + ("…" if len(name) > 40 else ""))
    bottom += h

if data_names:
    y_data = np.array([cf(data_names[0], k) for k in CUT_KEYS], dtype=float)
    ax.plot(x, y_data, color="k", marker="o", lw=1.5, label=f"Data ({data_names[0][:30]})")

ax.set_xticks(x)
ax.set_xticklabels(CUT_KEYS, rotation=25, ha="right")
ax.set_ylabel("Events")
ax.set_title("Cutflow: stacked MC + data")
ax.legend(bbox_to_anchor=(1.02, 1), loc="upper left", fontsize=7)
ax.set_yscale("log")
ax.grid(True, which="both", ls=":", alpha=0.5)
plt.tight_layout()
plt.show()
```
